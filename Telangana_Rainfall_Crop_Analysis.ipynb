```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load data
from google.colab import files
uploaded = files.upload()  # Upload your CSV file here

df = pd.read_csv("Telangana_Rainfall_Crop_Production.csv")
df.head()
```

```python
# Check summary
df.info()
df.describe()

# Sort by rainfall
df.sort_values(by="Rainfall_mm", ascending=False)

# Correlation between rainfall and production
df.corr(numeric_only=True)
```

```python
sns.barplot(x="District", y="Rainfall_mm", data=df)
plt.title("District-wise Rainfall in Telangana (2022)")
plt.xticks(rotation=45)
plt.show()

sns.scatterplot(x="Rainfall_mm", y="Rice_Production_Tonnes", data=df)
plt.title("Rainfall vs Rice Production")
plt.show()

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
##  Summary
- Warangal had the highest rainfall.
- Districts with more rainfall also saw higher rice production.
- Maize production seems less correlated with rainfall.
```
